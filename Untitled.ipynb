```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
data = pd.read_csv('movies.csv')

# Preprocess the data (handle missing values, encode categorical variables, select features)
data.dropna(inplace=True)  # Drop rows with missing values

# Feature engineering and selection
X = data[['budget', 'popularity', 'vote_count']]  # Select relevant features
# Additional feature engineering can be performed here, such as transforming or scaling features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Standardize the features

# Target variable
y = data['revenue']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Choose a regression model (e.g., Random Forest Regression)
model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
import requests

api_key = '649b944ed41440824b94f77a49a1e67d'
base_url = 'https://api.themoviedb.org/3'

endpoint = '/movie/popular'

# Adjust the number of pages to fetch more rows
pages_to_fetch = 50  # Fetching 3 pages (3 pages * 20 results per page = 60 rows)

# Initialize an empty DataFrame to store the results
df_combined = pd.DataFrame()

for page in range(1, pages_to_fetch + 1):
    # Additional parameters for discovering movies
    params = {
        'api_key': api_key,
        'sort_by': 'popularity.desc',
        'page': page
    }

    # Construct the URL for discovering movies
    url = f'{base_url}{endpoint}'

    # Make the API request for discovering movies
    response = requests.get(url, params=params)

    if response.status_code == 200:
        # Parse the JSON response for discovering movies
        movie_data = response.json()

        # Convert JSON to pandas DataFrame for the current page
        df_page = pd.json_normalize(movie_data['results'])

        # Append the current page's data to the combined DataFrame
        df_combined = df_combined.append(df_page, ignore_index=True)
    else:
        print(f"Error: {response.status_code}")

# Export the combined DataFrame to an Excel file
excel_filename = 'large_movie_data_combine.xlsx'
df_combined.to_excel(excel_filename, index=False)

print(f"Large movie dataset (50 rows) has been exported to {excel_filename}")
print("X")
```
